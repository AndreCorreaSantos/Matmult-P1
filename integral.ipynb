```python
import matplotlib.pyplot as plt
from math import *
import numpy as np

step = 0.0001

range_t = np.arange(0,pi*6,step)

def funcao(t):
    return sqrt(89-(80*((sin(t)*sin(8*t/3))+(cos(t)*cos(8*t/3)))))

lista_pontos = []
for i in range_t:
    lista_pontos.append(funcao(i))

plt.plot(lista_pontos)

def integra(lista,step):
    area = 0
    for i in range(0,len(lista)):
        if i != 0:
            area+= (lista[i-1] + lista[i])*step/2
    return area

integral = integra(lista_pontos,step)
print(integral)
print(len(lista_pontos))
```

```python
min_dist = 7.6113



def calcula_dist(p1,p2):
    return sqrt(((p1[0]-p2[0])**2)+((p1[1]-p1[1])**2))

def calcula_dist_centro(p1):
    return sqrt((p1[0]**2)+(p1[1]**2))

def funcao(t):
    return (5*cos(t)-3*cos(8*t/3),5*sin(t)-3*sin(8*t/3))

lista_dists_centro = []
lista_pontos2 = []

for i in range_t:
    if calcula_dist_centro(funcao(i)) >= min_dist:
        v = funcao(i)
        lista_pontos2.append(v)
        lista_dists_centro.append(calcula_dist_centro(v))

plt.plot(lista_dists_centro)

step2 = len(lista_pontos)*step/len(lista_dists_centro)

area_interna = integra(lista_dists_centro,step2)
print(area_interna)

```
